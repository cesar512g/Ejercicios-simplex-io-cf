```python
import numpy as np
from scipy.optimize import linprog

# Definir los coeficientes de la función objetivo
c = np.array([-2, -1])

# Definir las restricciones del problema
A = np.array([[2, 1]])
b = np.array([5])

# Definir los límites de las variables
x0_bounds = (0, None)
x1_bounds = (0, None)

# Resolver el problema usando linprog()
res = linprog(c, A_ub=A, b_ub=b, bounds=[x0_bounds, x1_bounds])


print("La funcion linprog se va a encargar de automizar el proceso simplex mostrando los pasos y el valor optimo","\n")

# Mostrar el resultado
print(res)

# Mostrar el valor óptimo de la función objetivo
print("valor optimo =  No hay valor que satisfaga este problema con la funcion objetivo planteada solucion incosistente")
```
